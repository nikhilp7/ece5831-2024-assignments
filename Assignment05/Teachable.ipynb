```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
from PIL import Image
import cv2
import os
```

```python
""" Step 1: Function to show an image using matplotlib """

def show_image(image_path):
    """Display the image using matplotlib."""
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
    plt.imshow(img_rgb)
    plt.axis('off')  # Hide axis
    plt.show()
```

```python
""" Testing the show_image function with a sample image path """

image_path = 'Rr.jpg'
```

```python
""" Test in the notebook environment """

show_image(image_path)
```

```python
""" Step 2: Testing command-line argument handling """

def parse_arguments(args):
    parser = argparse.ArgumentParser(description='Test parsing arguments.')
    parser.add_argument('--image-path', required=True, help='Path to the image to classify')
    return parser.parse_args(args)
```

```python
""" Example argument list (simulating command-line input) """

example_args = ['--image-path', 'Rr.jpg']
```

```python
""" Parse the arguments """

parsed_args = parse_arguments(example_args)
```
